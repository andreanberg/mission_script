```python
import numpy as np
from drone import Drone
from environment import Env
from simulator import Sim, Takeoff
from debug import Debugger
```

```python
drone = Drone(mass=6.0, wing_area=0.9, thrust=50.0)
env = Env(air_density=1.225)
sim = Sim(drone=drone, env=env, dt=0.01)

debug = Debugger(show=True, args=["v_body"])
to = Takeoff(runway_length=20, debugger=debug)
sim.run(to)

## TODO något script sådant att man kan se live med
# en plot som dynamiskt växer medan scriptet körs
# fixa problemet med var du är 
```
